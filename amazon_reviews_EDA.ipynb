## AMAZON REVIEWS EDA

This dataset contains information aboout electronic products listed on Amazon with their pricing details, discounts, customer ratings and reviews.

Each record represents a product and includes both quantitative and qualitative attributes related to product characteristics and user feedback.

#### Importing necessary libraries and dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("amazon.csv")
```

This dataset has 1465 records and 16 columns. There are two missing values in the dataset.

Column description:
- **product_id:** Unique identifier assigned to each product.
- **product_name:** Name or title of the product as listed on the platform.
- **category:** Product category represented in a hierarchical format, separated by the | symbol.
- **discounted_price:** Price of the product after discount (originally stored as text with currency symbols).
- **actual_price:** Original price of the product before discount.
- **discount_percentage:** Percentage discount applied to the product.
- **rating:** Average customer rating of the product on a scale of 1 to 5.
- **rating_count:** Total number of users who rated the product.
- **about_product:** Description of the product highlighting its features and specifications.
- **user_id:** Unique identifier(s) of users who reviewed the product.
- **user_name:** Name(s) of users who provided reviews.
- **review_id:** Unique identifier(s) for the reviews associated with the product.
- **review_title:** Title(s) of customer reviews.
- **review_content:** Textual content of customer reviews.
- **img_link:** URL linking to the product image.
- **product_link:** URL linking to the product page on the platform.

#### Data Preprocessing

```python
# delete columns with missing values
df.dropna()
```

```python
# standardize the column names 
df.columns = (df.columns.str.lower().str.strip().str.replace(" ", "_"))
df.columns
```

```python
# remove rupee symbol from 'discounted_price' and 'actual_price' columns
price_cols = ["actual_price", "discounted_price"]

for col in price_cols:
    df[col] = (df[col].str.replace("₹", "", regex=False).str.replace(",", "", regex=False).astype(float))
```

```python
# remove percentage symbol from 'discount_percentage'
df['discount_percentage'] = (df['discount_percentage'].str.replace("%", "", regex=False).astype(float))
```

```python
# remove comma symbol from 'rating_count'
df["rating_count"] = (df["rating_count"].astype(str).str.replace(",", "", regex=False).pipe(pd.to_numeric, errors="coerce").fillna(0).astype(int))
```

```python
# set ratings to float and in range(0 to 5) and remove rows with false mask
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")
mask = df["rating"].between(0, 5)
df = df[mask]
```

```python
# split into category and sub-categories
split_category = df["category"].str.split("|")

df["main_category"] = split_category.str[0]
df["product_type"] = split_category.str[-1]
df["sub_category"] = split_category.str[1:-1].str.join("|")
```

```python
# drop columns not required for EDA
df.drop(columns = ['product_id', 'category', 'about_product', 'user_id', 'user_name', 
           'review_id', 'review_title', 'img_link', 'product_link','sub_category', 'review_content'], inplace = True)
```

```python
# validate columns
df = df[df["discounted_price"] <= df["actual_price"]]

df["discounted_price"] = ((df["actual_price"] - df["discounted_price"]) / df["actual_price"]) * 100
```

### Exploratory Data Analysis

**Category Analysis**

```python
plt.figure(figsize=(10, 5))
ax = sns.countplot(data=df, x="main_category", order=df["main_category"].value_counts().index, palette="pastel")
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Product Count per Category")
plt.xlabel("Main Category")
plt.ylabel("Number of Products")
plt.xticks(rotation=45)
plt.show()
```

From this countplot, we can see that some categories like electronics, computers & accessories and home & kitchen dominate over other categories. Most of the products in the dataset belong to these categories.

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x="main_category", y="rating", data=df, palette="pastel")
plt.title("Rating Distribution by Category")
plt.xlabel("Main Category")
plt.ylabel("Rating")
plt.xticks(rotation=45)
plt.show()
```

Through this boxplot, we can see that most categories have ratings clustered between 4.0 and 4.5, meaning overall customer satisfaction is high across categories.

```python
avg_discount = (df.groupby("main_category")["discount_percentage"].mean().sort_values(ascending=False))

plt.figure(figsize=(10, 5))
ax = sns.barplot(x=avg_discount.index, y=avg_discount.values, palette="pastel")
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Average Discount Percentage by Category")
plt.xlabel("Product Type")
plt.ylabel("Average Discount (%)")
plt.xticks(rotation=90)
plt.show()
```

From the above barchart, we can see that Home Improvement, Computers & Acessories, Health & Personal Care and Electronics departments have an average discount of 50% and above. 

**Price and Discount Analysis**

```python
plt.figure(figsize=(12, 6))
ax = sns.histplot(df["actual_price"], bins=20, kde=True)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Distribution of Actual Product Prices")
plt.xlabel("Actual Price")
plt.ylabel("Product Count")
plt.show()
```

Most products are concentrated in the lower price range, indicating a budget-oriented marketplace.

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["discount_percentage"], bins=20, kde=True)
plt.title("Distribution of Discount Percentage")
plt.xlabel("Discount Percentage")
plt.ylabel("Product Count")
plt.show()
```

Moderate discounts (40–65%) are most common, while extreme discounts are relatively rare.

**Rating and Popularity**

```python
top_reviewed = df.sort_values(by="rating_count",ascending=False).head(11)
plt.figure(figsize=(9, 7))
ax = sns.barplot(data=top_reviewed, x="rating_count", y="product_name", palette="pastel")
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Top 10 Most Reviewed Products")
plt.xlabel("Rating Count")
plt.ylabel("Product Name")
plt.show()
```

A small number of products attract a disproportionate number of reviews.

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["rating"], bins=20, kde=True)
plt.title("Distribution of Product Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Products")
plt.show()
```

Most of the products have high ratings between 4 and 4.5.

```python
numeric_cols = ["discounted_price", "actual_price", "discount_percentage", "rating", "rating_count"]
corr = df[numeric_cols].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="Blues", linewidths=0.5)
plt.title("Correlation Heatmap of Price, Rating & Popularity")
plt.show()
```

The correlation analysis reveals that pricing and discounts have minimal influence on customer ratings. 

While discounted price and discount percentage are perfectly correlated due to their derived nature, both show a weak negative relationship with ratings, indicating that cheaper products tend to receive slightly higher ratings. 

Actual price and rating count show negligible correlations with ratings, suggesting that neither premium pricing nor product popularity strongly impacts customer satisfaction.
